```python
import matplotlib.pyplot as plt
from pymystem3 import Mystem
import pandas as pd
import re
from string import punctuation

mystem = Mystem()
```

```python
def extract_slot_X(pattern, construction, text): # нижний регистр + убрать пунктуацию
  text = " ".join([token.lower() for token in text.split()])
  collocation = re.search(pattern, text).group(0)
  return re.sub(construction, "", collocation)
```

```python
df_1 = pd.read_csv("vplot_do_X.csv")
df_1
```

```python
res_1 = [extract_slot_X(r"вплоть до [^ ]+", r"вплоть до", example) for example in df_1["Пример"].replace(r'[^\w\s]',' ',regex=True).tolist()]
lemmas_1 = [lemma for lemma in mystem.lemmatize(" ".join(res_1)) if lemma != ' ' and lemma != '\n' and lemma != '  ']
res_1_dict = {i:lemmas_1.count(i) for i in lemmas_1}
res_1_dict
```

```python
pd.DataFrame.from_dict(res_1_dict, orient='index')
```
